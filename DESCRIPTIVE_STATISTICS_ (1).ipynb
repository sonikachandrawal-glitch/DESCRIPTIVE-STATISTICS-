Q.1  Understanding Central Tendency
A bakery tracks the daily sales of muffins (in dozens) over a week: [10, 12, 11, 15, 14, 13, 12].

    10+12+11+15+14+13+12 = 87

    87 / 7 = 12.43 DOZENS

     1. THE DATA IS CLOSE TOGETHER (NO EXTREME HIGH VALUES OR LOW VALUES )

       PICKING THE HIGHEST OR LOWEST .
